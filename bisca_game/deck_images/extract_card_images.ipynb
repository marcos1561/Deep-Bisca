```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
suits = ["O", "C", "E", "P"]

names = []
for s in suits:
    for i in range(1, 13):
        names.append(f"{i}-{s}")

idx_to_names = dict(zip(range(12*4 + 2), names + ["CORINGA", "CORINGA2"]))

idx_to_names
```

```python
img = mpimg.imread("deck_image_complete.jpg")

img_shape = img.shape
img = img[3:-4,1:-3]
img_shape = img.shape

card_shape = (222, 146)

h_sep_length = 4 
v_sep_length = 2

count = 0
col = 0
while col+card_shape[1] <= img_shape[1]:
    row = 0
    row_count = 0
    while row+card_shape[0] <= img_shape[0]:
        card_img = img[row:row+card_shape[0],col:col+card_shape[1]]
        plt.title(idx_to_names[count])

        plt.imshow(card_img)
        plt.imsave("..\\cards_images2\\" + idx_to_names[count] + ".jpg", card_img)
        plt.show()

        row += card_shape[0] + v_sep_length
        row_count += 1
        count += 1
    if row_count != 5:
        print("ERRO")
    col += card_shape[1] + h_sep_length
```
